🛠️ חלק 6 – בניית Pipeline ו-ColumnTransformer

```python
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns

print("Numeric features:", numeric_features)
print("Categorical features:", categorical_features)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
numeric_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_pipeline, numeric_features),
        ('cat', categorical_pipeline, categorical_features)
    ]
)
```

```python
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)
```

### בניית Pipeline לעיבוד נתונים

בשלב זה נבנה Pipeline לעיבוד הנתונים לפני אימון המודלים.
המשתנים הנומריים עברו השלמת ערכים חסרים באמצעות חציון
ולאחר מכן Scaling.
המשתנים הקטגוריאליים עברו השלמת ערכים חסרים באמצעות הערך השכיח
וקידוד באמצעות One-Hot Encoding.

כל שלבי העיבוד בוצעו באמצעות ColumnTransformer,
כאשר תהליך ה-fit בוצע רק על סט האימון
כדי למנוע זליגת מידע.


### שאלות תיאורטיות – Pipelines ו-Preprocessing

**מדוע מודלים דורשים קלט נומרי בלבד?**  
מרבית מודלי למידת המכונה מבצעים חישובים מתמטיים,

🔄 חלק 7 – אימון מודלים והשוואה
אנחנו נאמן:
1️⃣ Logistic Regression
2️⃣ KNN
3️⃣ Decision Tree

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
log_reg = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', LogisticRegression(max_iter=1000))
])

log_reg_params = {
    'model__C': [0.01, 0.1, 1, 10]
}

log_reg_grid = GridSearchCV(
    log_reg,
    log_reg_params,
    cv=5,
    scoring='accuracy'
)

log_reg_grid.fit(X_train, y_train)
```

```python
knn = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', KNeighborsClassifier())
])

knn_params = {
    'model__n_neighbors': [3, 5, 7, 9]
}

knn_grid = GridSearchCV(
    knn,
    knn_params,
    cv=5,
    scoring='accuracy'
)

knn_grid.fit(X_train, y_train)
```

```python
tree = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', DecisionTreeClassifier(random_state=42))
])

tree_params = {
    'model__max_depth': [None, 5, 10, 20],
    'model__min_samples_split': [2, 5, 10]
}

tree_grid = GridSearchCV(
    tree,
    tree_params,
    cv=5,
    scoring='accuracy'
)

tree_grid.fit(X_train, y_train)
```

```python
import pandas as pd

results = pd.DataFrame({
    'Model': ['Logistic Regression', 'KNN', 'Decision Tree'],
    'Best CV Accuracy': [
        log_reg_grid.best_score_,
        knn_grid.best_score_,
        tree_grid.best_score_
    ]
})

results
```

### אימון מודלים והשוואתם

בשלב זה אומנו שלושה מודלים שונים באמצעות GridSearchCV:
Logistic Regression, KNN ו-Decision Tree.
לכל מודל נבחרו היפר-פרמטרים מתאימים,
והאימון בוצע באמצעות Cross-Validation עם cv=5.

השוואת המודלים בוצעה על סמך דיוק ממוצע ב-Cross-Validation,
כדי לבחור את המודל בעל הביצועים הטובים ביותר.

False Positive הוא מצב שבו המודל חזה חיובי אך הערך האמיתי שלילי,
ואילו False Negative הוא מצב שבו המודל חזה שלילי אך הערך האמיתי חיובי.

**מתי Recall חשוב יותר מ-Precision?**  
Recall חשוב יותר כאשר החמצת מקרים חיוביים
מהווה בעיה חמורה יותר מאשר זיהוי שגוי,
לדוגמה במצבים רפואיים או זיהוי סיכונים.
